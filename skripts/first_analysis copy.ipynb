```python
import pandas as pd

def to_number_or_none(val):
    """
    Try to convert `val` to a float.
    • Handles strings like "1,23" by switching comma→dot.
    • Returns None if conversion fails.
    """
    if pd.isna(val):          # already NaN/None
        return None
    try:
        # convert comma‐decimal strings → dot
        if isinstance(val, str):
            val = val.replace(",", ".")
        return float(val)
    except (ValueError, TypeError):
        return None

# apply to every cell
df = df.applymap(to_number_or_none)
```

```python
import re

# Define a mapping from JSON special characters to UTF-8 alternatives
json_to_utf8 = {
    '"': '',   # left double quotation mark
    "'": '',   # left single quotation mark
    '{': '',  # fullwidth left curly bracket
    '}': '',  # fullwidth right curly bracket
    '[': '',  # left black lenticular bracket
    ']': '',  # right black lenticular bracket
    ':': '',  # fullwidth colon
    ',': '',  # fullwidth comma
    '\\': '', # fullwidth reverse solidus
    " ": ""
}

def replace_json_chars(col):
    # Replace each JSON character with its UTF-8 alternative
    for k, v in json_to_utf8.items():
        col = col.replace(k, v)
    return col

final_df.columns = [replace_json_chars(col) for col in final_df.columns]
```

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.model_selection import LeaveOneOut, cross_val_predict
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# ──────────────────────────────────────────────────────────────────
# 1 DATA PREP
# ──────────────────────────────────────────────────────────────────
target   = "ecoli"
drop_out = [target, "entro", 
            "pos_neg"]               # obvious leaks

df = df.copy()                                          # your DataFrame
df = df.dropna(subset=[target])

# coerce numeric-looking strings
for col in df.select_dtypes("object"):
    df[col] = pd.to_numeric(df[col].str.replace(",", "."), errors="coerce")

X_full = df.drop(columns=drop_out)
y      = df[target]

numeric_cols = (
    X_full.select_dtypes(np.number)
          .loc[:, lambda d: d.nunique(dropna=True).gt(1)]   # non-constant
          .dropna(axis=1, how="all")                        # not all-NaN
          .columns.tolist()
)
if not numeric_cols:
    raise ValueError("No usable numeric predictors left.")

# ──────────────────────────────────────────────────────────────────
# 2 HELPER: Ridge + LOOCV metrics
# ──────────────────────────────────────────────────────────────────
def ridge_loocv_metrics(X_df, y_ser, alpha=1.0):
    """Return MAE, RMSE, R² and an unfitted pipeline template."""
    pipe = Pipeline([
        ("imp",   SimpleImputer(strategy="mean")),
        ("scale", StandardScaler()),
        ("model", Ridge(alpha=alpha)),
    ])
    loo    = LeaveOneOut()
    y_pred = cross_val_predict(pipe, X_df, y_ser, cv=loo)
    mae  = mean_absolute_error(y_ser, y_pred)
    rmse = mean_squared_error(y_ser, y_pred, squared=False)
    r2   = r2_score(y_ser, y_pred)
    return mae, rmse, r2, pipe            # pipe is still unfitted

# ──────────────────────────────────────────────────────────────────
# 3 RFE LOOP
# ──────────────────────────────────────────────────────────────────
curr_feats = numeric_cols.copy()
history    = []    # (n_feats, mae, rmse, r2, feats)

best_rmse  = np.inf
best_feats = None

round_no = 1
while len(curr_feats) >= 2:
    X_curr = X_full[curr_feats]
    mae, rmse, r2, pipe_tpl = ridge_loocv_metrics(X_curr, y)

    history.append((len(curr_feats), mae, rmse, r2, curr_feats.copy()))
    print(f"Round {round_no:>2}: feats={len(curr_feats):>3}  "
          f"MAE={mae:8.1f}  RMSE={rmse:8.1f}  R²={r2:7.3f}")

    if rmse < best_rmse:
        best_rmse  = rmse
        best_feats = curr_feats.copy()

    # fit once to rank coefficients, drop the least important
    pipe_fitted = pipe_tpl.fit(X_curr, y)
    coef_abs    = np.abs(pipe_fitted.named_steps["model"].coef_)
    drop_feat   = curr_feats[int(np.argmin(coef_abs))]
    curr_feats.remove(drop_feat)
    print(f"          ➜ dropped '{drop_feat}' (|coef|={coef_abs.min():.4g})\n")
    round_no += 1

# ──────────────────────────────────────────────────────────────────
# 4 BEST SUBSET METRICS
# ──────────────────────────────────────────────────────────────────
print("\n🏆  BEST SUBSET BASED ON LOOCV RMSE")
print(f"   • features used : {len(best_feats)}")
mae_best, rmse_best, r2_best, _ = ridge_loocv_metrics(X_full[best_feats], y)
print(f"   • MAE           : {mae_best:.1f}")
print(f"   • RMSE          : {rmse_best:.1f}")
print(f"   • R²            : {r2_best:.3f}")

# ──────────────────────────────────────────────────────────────────
# 5 FINAL REFIT & COEFFICIENT PLOT
# ──────────────────────────────────────────────────────────────────
pipe_best = Pipeline([
    ("imp",   SimpleImputer(strategy="mean")),
    ("scale", StandardScaler()),
    ("model", Ridge(alpha=1.0)),
]).fit(X_full[best_feats], y)

ridge_best = pipe_best.named_steps["model"]
coefs      = ridge_best.coef_

imp = (pd.DataFrame({"feature": best_feats, "coef": coefs})
       .assign(abs_coef=lambda d: d.coef.abs())
       .sort_values("abs_coef", ascending=False))

print("\nTop coefficients for best model:")
print(imp.head(10))

plt.figure(figsize=(10, 6))
bars = ["green" if c > 0 else "red" for c in imp.coef]
plt.barh(imp.feature, imp.coef, color=bars)
plt.axvline(0, color="black", lw=0.8)
plt.title(f"Ridge coefficients ({len(best_feats)} features)")
plt.xlabel("Coefficient")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
